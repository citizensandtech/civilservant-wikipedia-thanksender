# Goal
create an account map beteween:
    user_name <-> public_anonymous_id <-> private_anonymous id


```python
import pandas as pd
import os
import json
import uuid
```

```python
datadir = '/data/project/frwelcome/datasets/survey data/'
survey_sent_f = 'pilot_experiment_actions_March_12_2020.csv'
f = os.path.join(datadir, survey_sent_f)
```

```python
df = pd.read_csv(f , sep='\t') 
```

```python
df['metadata_json'] = df['metadata_json'].apply(lambda s: json.loads(s))
```

```python
df['public_anonymous_id'] = df['metadata_json'].apply(lambda d: d['public_anonymous_id'])
df['user_name'] = df['metadata_json'].apply(lambda d: d['user_name'])
df['lang'] = df['metadata_json'].apply(lambda d: d['lang'])
```

```python
namespace = uuid.uuid4()
```

```python
df['private_anonymous_id'] = df['user_name'].apply(lambda u: uuid.uuid3(namespace, u)) 
```

```python
df['user_name'].plot(kind='hist')
```

```python
df.plot()
```

```python
acct_map_f = os.path.join(datadir, 'pilot_account_mapping.csv')
df[['lang','user_name','public_anonymous_id','private_anonymous_id']].to_csv(acct_map_f, index=False)
```
